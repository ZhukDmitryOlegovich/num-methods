```python
import pandas as pd
```

```python
loc1 = pd.read_csv("loc1.txt", sep=" ")
loc2 = pd.read_csv("loc2.txt", sep=" ")
loc3 = pd.read_csv("loc3.txt", sep=" ")
```

```python
def analyze_loc(loc):
	print("head:", loc.head())
	print("info:", loc.info())
	print("describe:", loc.describe())
	print("isnull:", loc.isnull())
```

```python
analyze_loc(loc1)
```

```python
analyze_loc(loc2)
```

```python
analyze_loc(loc3)
```
